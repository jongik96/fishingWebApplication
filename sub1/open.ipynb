```python
import pandas as pd
pd.set_option('display.max_columns',None)
data=pd.read_csv("./갯바위.csv")
data
```

```python
import re
fish=data.TARGET.str.findall(r'▶\s\S*\s|▶\S*\s')
fish
```

```python
arr=[]
for i in fish:
  temp=[]
  for j in i:
    j=j.replace('▶','').strip()
    temp.append(j)
  arr.append(temp)
arr
```

```python
data['fish']=arr
data
```

```python
s=",".join(data['fish'][0])
s
```

```python
newarr=[]
for i in data['fish']:
  s=",".join(i)
  newarr.append(s)
newarr

  
```

```python
data['FISH']=newarr
data
```
