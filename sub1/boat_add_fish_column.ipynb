```python
# fish parsing 하기
import pandas as pd
pd.set_option('display.max_columns',None)
data=pd.read_csv("./선상.csv")
data
```

```python
import re
fish=data.TARGET.str.findall(r'▶\s\S*\s|▶\S*\s')
fish
```

```python
arr=[]
for i in fish:
  temp=[]
  for j in i:
    j=j.replace('▶','').strip()
    temp.append(j)
  arr.append(temp)
arr
```

```python
newarr=[]
for i in arr:
  s=",".join(i)
  newarr.append(s)
newarr
```

```python
data['FISH']=newarr
data
```

```python
data.to_csv('new선상.csv', index=False)
```
