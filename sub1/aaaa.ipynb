```python
import numpy as np
import pandas as pd
import json
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
meta = pd.read_csv("./fishing.csv")
meta = meta.rename(columns={'ID':'fishingId'})
meta.head()
```

```python
ratings = pd.read_csv("./reviewdata.csv")


ratings = ratings[['id', 'userId', 'fishingId', 'rating']]
ratings = ratings.rename(columns={'id':'reviewId'})

ratings.head()
```

```python
data = pd.merge(ratings, meta, on='fishingId', how='inner') 
data.head()
```

```python
matrix = data.pivot_table(index='userId', columns='POINT_NM', values='rating')
matrix.head(20)
```

```python
FISHING_WEIGHT = 0.1

def pearsonR(s1, s2):
    s1_c = s1 - s1.mean()
    s2_c = s2 - s2.mean()
    return np.sum(s1_c * s2_c) / np.sqrt(np.sum(s1_c ** 2) * np.sum(s2_c ** 2))

def recommend(input_POINT_NM, matrix, n, similar_genre=True):
    input_FISH = meta[meta['POINT_NM'] == input_POINT_NM]['FISH'].iloc(0)[0]

    result = []
    for title in matrix.columns:
        if title == input_POINT_NM:
            continue

        # rating comparison
        cor = pearsonR(matrix[input_POINT_NM], matrix[title])
        
        # genre comparison
        if similar_genre and len(input_FISH) > 0:
            temp_FISH = meta[meta['POINT_NM'] == title]['FISH'].iloc(0)[0]
            same_count = np.sum(np.isin(input_FISH, temp_FISH))
            cor += (FISHING_WEIGHT * same_count)

        if np.isnan(cor):
            continue
        else:
            # 위치 기반 정보 넣어주고 cor에 더해주기(추후 작업)
            
            result.append((title, '{:.2f}'.format(cor), temp_FISH))
        

            
    result.sort(key=lambda r: r[1], reverse=True)
    print(result[:n])
    return result[:n]

recommend_result = recommend('거북바위', matrix, 3, similar_genre=True)
pd.DataFrame(recommend_result, columns = ['POINT_NM', 'Correlation', 'FISH'])

```

```python

# lst는 실제로 본인이 평점 좋게 준 낚시터들 상위 3개를 가지고 옴

lst = ['달천도 부근', '낭도산 동측', '거북바위']
res = []
for i in range(len(lst)):
    recommend_result = recommend(lst[i], matrix, 3, similar_genre=True)
    res.append(pd.DataFrame(recommend_result, columns = ['POINT_NM', 'Correlation', 'FISH']))
    
abc = pd.concat(res)
abc.sort_values(by=['Correlation'], axis=0, ascending=False)
     
```
