```python
import numpy as np
import pandas as pd
import json
```

```python
meta = pd.read_csv('fish-dataset/fishing.csv')

meta.head()
```

```python
meta = meta[['ID', 'NAME', 'POINT_NM', 'ADR_KNM', 'CATEGORY']]
meta = meta.rename(columns={'ID':'fishingId'})

meta.head()
```

```python
ratings = pd.read_csv('fish-dataset/Reviewdata.csv')
ratings = ratings[['userId', 'fishingId', 'rating']]
ratings.head()
```

```python
ratings.describe()
```

```python
meta.fishingId = pd.to_numeric(meta.fishingId, errors='coerce')
ratings.fishingId = pd.to_numeric(ratings.fishingId, errors='coerce')
```

```python
data = pd.merge(ratings, meta, on='fishingId', how='inner')

data.head()
```

```python
matrix = data.pivot_table(index='userId', columns='POINT_NM', values='rating')
matrix.head(20)
```

```python
GENRE_WEIGHT = 0.1

def pearsonR(s1, s2):
    s1_c = s1 - s1.mean()
    s2_c = s2 - s2.mean()
    return np.sum(s1_c * s2_c) / np.sqrt(np.sum(s1_c ** 2) * np.sum(s2_c ** 2))

def recommend(input_fishing, matrix, n, similar_fishing=True):
    input_fishing = meta[meta['POINT_NM'] == input_fishing]['CATEGORY'].iloc(0)[0]

    result = []
    for title in matrix.columns:
        if title == input_fishing:
            continue

        cor = pearsonR(matrix[input_fishing], matrix[title])

        if similar_fishing and len(input_fishing) > 0:
            temp_fishing = meta[meta['POINT_NM'] == title]['CATEGORY'].iloc(0)[0]

            same_count = np.sum(np.isin(input_fishing, temp_fishing))
            cor += (GENRE_WEIGHT * same_count)
        
        if np.isnan(cor):
            continue
        else:
            result.append((title, '{:.2f}'.format(cor), temp_fishing))
            
    result.sort(key=lambda r: r[1], reverse=True)

    return result[:n]
```

```python
for title in matrix.columns:
    print(matrix[title])
```

```python
recommend_result = recommend('가란도', matrix, 10, similar_fishing=True)

pd.DataFrame(recommend_result, columns = ['Title', 'Correlation', 'Genre'])
```
